### Data Cleaning

### To be cleaned
- Account ID is string.
- Last Year Growth Indicator is category.
- Account Class is category.
- convert account class = 0 to nan.
- Profitability = -999.99 or 100 is weird and usual so i preferred to convert it to nan not to drop because it has important observation in historical revenues.
- i have doubts about the records of observations which are nan in Last Year Growth Indicator and Profitability and 0 in Historical Revenue so i preferred to convert their Historical Revenue to nan.
- convert Total open opportunities_USD = 0  to nan as it unlogic.
- drop the account id = 12343 as it has not data to be explored.

```python
df_clean = df.copy()
```

```python
# Account ID is string.
df_clean['Account ID'] = df_clean['Account ID'].astype('str')

# Last Year Growth Indicator is category.
df_clean['Last Year Growth Indicator'] = df_clean['Last Year Growth Indicator'].astype('category')

9# Account Class is category.
df_clean['Account Class'] = df_clean['Account Class'].astype('category')

df_clean.info()
```

```python
# convert account class = 0 to nan.
df_clean.loc[df_clean['Account Class'] == 0, 'Account Class'] = np.nan

# convert Profitability = -999.99 or 100 to nan.
df_clean.loc[df_clean['Profitability'] == 100, 'Profitability'] = np.nan
df_clean.loc[df_clean['Profitability'] == -999.99, 'Profitability'] = np.nan

# convert Historical Revenue = 0 to nan.
df_clean.loc[df_clean['Historical Revenue'] == 0, 'Historical Revenue'] = np.nan

# convert Total open opportunities_USD = 0  to nan.
df_clean.loc[df_clean['Total open opportunities_USD'] == 0, 'Total open opportunities_USD'] = np.nan

# drop the account id = 12343 as it has not data to be explored.
df_clean.drop(df_clean.index[79], axis=0, inplace = True)
```
